```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
from numpy.fft import *
import math
import wave
import struct
from scipy.io import wavfile
```

```python
#SF = Sampling frequency
#SD = Signal Data
SD, SF = librosa.load('/home/ncbc/Desktop/WorkSpace_NCBC/Siren/spectrograms_final audio/Graph 3.wav',sr=None)
```

```python
SF
```

```python
SD
```

```python
window_size = 1024
hop_length = 512
n_mels = 128
time_steps = 384
```

```python
window = np.hanning(window_size)
stft= librosa.core.spectrum.stft(SD, n_fft = window_size, hop_length = hop_length, window=window)
out = 2 * np.abs(stft) / np.sum(window)
out
```

### Frequency vs Amplitude

```python
plt.figure(figsize=(12, 4))
ax = plt.axes()
librosa.display.specshow(librosa.amplitude_to_db(out, ref=np.max), y_axis='time', x_axis='log',sr=SF )
plt.xlabel('Frequency[Hz]')
plt.ylabel('Amplitude')
#plt.title('Two-Tone')
plt.set_cmap('gist_heat')
plt.savefig('frequency_vs_Amplitude.png',dpi=600, transparent=True)
plt.show()
```

### Time vs Frequency

```python
plt.figure(figsize=(12, 4))
ax = plt.axes()
librosa.display.specshow(librosa.amplitude_to_db(out, ref=np.max), y_axis='log', x_axis='time',sr=SF )
plt.xlabel('Time [s]')
plt.ylabel('Frequency')
#plt.title('Two-Tone')
plt.set_cmap('gist_heat')
plt.savefig('Time_vs_Freq.png',dpi=600, transparent=True)
plt.show()
```

### Amplitude vs Time

```python
plt.figure(figsize=(12, 4))
ax = plt.axes()
librosa.display.specshow(librosa.amplitude_to_db(out, ref=np.max), y_axis='mel', x_axis='time',sr=SF )
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Two-Tone')
plt.set_cmap('gist_heat')
plt.savefig('Amplitude_vs_time.png',dpi=600, transparent=True)
plt.show()
```
