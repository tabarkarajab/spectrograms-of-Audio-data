# **Importing Libraries**

```python
#importing libraries for audio spectrograms
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
from numpy.fft import *
import math
import wave
import struct
from scipy.io import wavfile
```

```python
#SF = Sampling frequency
#SD = Signal Data
SD, SF = librosa.load('../input/ambulance-audio-dataset/Graph 3.wav',sr=None)
```

```python
SF
```

```python
SD
```

```python
window_size = 1024
hop_length = 512
n_mels = 128
time_steps = 384
```

```python
window = np.hanning(window_size)
stft= librosa.core.spectrum.stft(SD, n_fft = window_size, hop_length = hop_length, window=window)
out = 2 * np.abs(stft) / np.sum(window)
```

# Visualizing the Audio Sounds in different color mapping


```python
plt.figure(figsize=(12, 4))
ax = plt.axes()
librosa.display.specshow(librosa.amplitude_to_db(out, ref=np.max), y_axis='log', x_axis='time',sr=SF )
plt.ylabel('Frequency[Hz]')
plt.xlabel('Time[sec]')
plt.title('Two-Tone')
plt.savefig('graph1_yelp.png', dpi=600, transparent=True)
plt.show()
```
